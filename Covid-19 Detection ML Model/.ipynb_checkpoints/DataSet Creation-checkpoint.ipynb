```python
import pandas as pd
import os
import numpy as np

datapath1='path\to\dataset'
dataset_path='path\to\store\the\output'

```

```python
categories=os.listdir(dataset_path)
print(categories)

dataset=pd.read_csv('covid-chestxray-dataset-master\\metadata.csv')
findings=dataset['finding']
print(findings)
image_names=dataset['filename']
print(image_names)
```

```python
positives_index=np.concatenate((np.where(findings=='Pneumonia/Viral/COVID-19')[0],np.where(findings=='Pneumonia/Viral/SARS')[0]))
print(len(positives_index))
positive_image_names=image_names[positives_index]
```

```python
import cv2

for positive_image_name in positive_image_names:
    image=cv2.imread(os.path.join(datapath1,'images',positive_image_name))
    try:
        cv2.imwrite(os.path.join(dataset_path,categories[2],positive_image_name),image)
    except Exception as e:
        print(e)
```

```python
datapath2='path\\to\\Coronahack-Chest-Xray-Dataset'

dataset=pd.read_csv('Chest_xray_Corona_Metadata.csv')
print(dataset)
findings=dataset['Label']
print(findings)
image_names=dataset['X_ray_image_name']
```

```python
negative_index=np.where(findings=='Normal')[0]
print(negative_index)
```

```python
negative_image_names=image_names[negative_index]
```

```python
import cv2
for negative_image_name in negative_image_names:
    image=cv2.imread(os.path.join(datapath2,'images',negative_image_name))
    try:
        cv2.imwrite(os.path.join(dataset_path,categories[1],negative_image_name),image)
    except Exception as e:
        print(e)
```
